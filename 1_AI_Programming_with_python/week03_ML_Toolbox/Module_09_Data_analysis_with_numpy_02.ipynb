# ndarray Manipulation 1

```python
import numpy as np
arr = np.random.randint(1, 100, size= (10, 5))
print(arr)
```

```python
print(arr.shape)
```

```python
b = arr.reshape(5, 10)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# row wise flattning
flat_b = b.flatten()
print(b)
print(b.ndim)
ravel_b = np.ravel(b, order='C')   
print(ravel_b)
```

```python
# column wise flattning 
ravel_b = np.ravel(b, order='F')
print(ravel_b)
```

## concat array

```python
a = np.random.randint(1, 10, size= (2,3))
b = np.random.randint(20, 30, size= (2,3))
print(a)
print(b)
```

```python
# row wise  (column need to be same size)
con_row = np.concatenate((a,b), axis = 0)
print(con_row)
```

```python
# column wise (row need to be same size)
con_column = np.concatenate((a,b), axis=1)
print(con_column)
```

# ndarray Arithmatic Operation and mathematical function

## arithmatic function and mathematical function

```python
# arithmatic 
x = np.array([10 ,20, 30, 40, 50, 60])
y = np.array([2, 4, 6, 8, 10, 11])
z = np.array([2, 4, 6, 8, 10, 11])
add = x+y+z # add the consecuennt elements of two arrays (arrays need to be in same size)
sub = x-y
mult = x*y*z
div = x/y
mod = x%y
com = x-(y+z)*x
print(add)
print(sub)
print(mult)
print(div)
print(mod)
print(com)

```

```python
# trigonometric
a = np.sin(np.deg2rad(60)) #for single number the fucntion takes the input in radian 
print(a)

sin_x = np.sin(np.deg2rad(x))
print(sin_x)
cos_y = np.cos(y)
print(cos_y)
```

```python
log = np.log10(x) # 10 based log
print(log)
print(np.log2(x))  # 2 based log
print(np.sqrt(x))
```

# Numpy Logical Function

```python
# comparison >, <, ==, >=, <=, !=
x = np.array([10 ,20, 30, 40, 50, 60])
y = np.array([2, 40, 6, 8, 50, 11])

greater = x>=y
print(greater)
equal = x==y
print(equal)
print(np.all(x==y))
print(np.any(x==y))
```

# ndarray sort

```python

x = np.array([11,  1, 15,  7, 28, 12, 13, 26, 24, 17,  6, 24])
print(x)
```

```python
# inplace sorting
z = x.copy()
print(z)
z.sort()
print(z)
```

```python
# copy sort
print(np.sort(x))  
print(x)
```

```python
# 2d array soritng
mat = np.array([[16, 20, 12, 20],
       [38, 33, 21, 18],
       [37, 12, 25, 35],
       [32, 26, 39, 31]])
print(mat)
```

```python
print(np.sort(mat, axis=0)) # vertical sort
print(np.sort(mat, axis=1)) # horizontal sort
```

# Searching

```python
print(x)
print(np.where(x ==12))  # np.where return the indics of the values
print(np.where(x >3)) # np.where return the indics of the values

# np.where(condition, x, y) if true it be x else it will be y
print(np.where(x>12, x, 0)) # here if x>12 it will remain x else turn into 0
print(x)
```

```python
# 2d searching
print(mat)
print(np.where(mat < 16))
print((np.where(mat < 16, "Failed", "Passed")))
```

```python
# min max
print(x)
print(np.argmax(x))
print(np.argmin(x))
```

# Counting 

```python
a = np.random.randint(50, 100, size= (100))
print(a)
```

```python
cnt = np.count_nonzero(a>76)
print(cnt)
cnt = np.count_nonzero(a==61)
print(cnt)
cnt = np.count_nonzero(a<68)

print(cnt)
```

```python
print(np.unique(a)) #print unique values
unq_val, unq_cnt = np.unique(a, return_counts=True) # return unique values and 
print(unq_val)
print(unq_cnt)
```

# Stastical Function

```python
# data Load
data = np.genfromtxt('student_marks_50.csv', delimiter=',', skip_header=1)
data
```

```python
math_marks = data[::, 0:1]
print(math_marks.T)
mx_math_marks = np.max(math_marks)
print(mx_math_marks)
mn_math_marks = np.min(math_marks)
print(mn_math_marks)
median_math_marks = np.median(math_marks)
print(median_math_marks)
std_math_marks = np.std(math_marks)
print(std_math_marks)
```

## correlation

```python
study_hour = np.array([2,3,4,5,6,7])
final_marks = np.array([50,60,70, 75, 80, 90])
data = np.array([study_hour, final_marks])
print(data)
cor = np.corrcoef(data)
print(cor)
```

# Linear Algebra

```python
a = np.array([
    [1,2,3],
    [4,5,6]
])
b = np.array([
    [2,3],
    [7,8],
    [4, 9]
])
sq = np.array([
    [1,2,3],
    [2,3,4],
    [4,5,6]
])
dot_product = np.dot(a,b)
print(dot_product)
print(np.trace(sq))

# derterminant
print(np.linalg.det(sq))
# rank
print(np.linalg.matrix_rank(sq))
```
