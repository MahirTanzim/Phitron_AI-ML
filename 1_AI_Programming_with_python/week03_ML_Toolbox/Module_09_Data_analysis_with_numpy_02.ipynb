```python
import numpy as np
arr = np.random.randint(1, 100, size= (10, 5))
print(arr)
```

```python
print(arr.shape)
```

```python
b = arr.reshape(5, 10)
print(b)
print(b.shape)
print(b.ndim)
```

```python
# row wise flattning
flat_b = b.flatten()
print(b)
print(b.ndim)
ravel_b = np.ravel(b, order='C')   
print(ravel_b)
```

```python
# column wise flattning 
ravel_b = np.ravel(b, order='F')
print(ravel_b)
```
