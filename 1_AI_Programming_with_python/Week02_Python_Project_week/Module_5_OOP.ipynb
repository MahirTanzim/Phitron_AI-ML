# Class and Object

```python
# Base class / Parent class
class Phone:
    category= 'Electronics'
    ## constructor
    def __init__(self, model, battery, camera, battery_percentage=100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_percentage = battery_percentage

    # methods
    def charge(self, hour):
        self.charge+=hour
    def capture(self, total_photo):
        if(self.battery_percentage <= 0):
            print("NO Charge")
        else:
            self.battery_percentage-=total_photo
            print(f"photo captured in {self.model}")
```

```python
## creating object
apple = Phone('Iphone 19', '5000maH', "48MP")
blueberry = Phone('B-12', "6000maH", "64MP")
motorola = Phone('M12', "5500maH", "12MP")
apple.category = "Super Electronics"
# print(apple)
# print(type(apple))
print(apple.category)
print(blueberry.category)
print(motorola.category)

print(apple.camera)
print(motorola.battery )
print(motorola.battery_percentage)
```

```python
print(apple.battery_percentage)
apple.capture(20)
print(apple.battery_percentage)
```

# Single Inheritance

```python
# child class / derived class
class Smartphone(Phone):
    def __init__(self, model, battery, camera, processor):
        super().__init__(model, battery, camera)
        self.processor = processor

    def charge(hour):
        print("Fast Charging in process")
        super().charge(hour)
```

```python
pro = Smartphone('X', 5000, 100, "Snapdragon")
print(pro.processor )
print(pro.battery_percentage)
pro.capture(20)
print(pro.battery_percentage)
```
