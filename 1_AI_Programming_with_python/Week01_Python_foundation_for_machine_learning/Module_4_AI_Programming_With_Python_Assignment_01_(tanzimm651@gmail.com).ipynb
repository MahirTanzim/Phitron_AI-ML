# Solution 3

```python
with open("./sample_data/data.txt", "r") as file:
    content = file.read()
    content = content.lower()
    # print(content)
    word = content.split()
    frq ={}
    for w in word:
        if w in frq:
            frq[w]+=1
        else:
            frq[w]=1
    # print(frq)
    ans = ''
    mx = 0
    for i in frq:
        if frq[i] > mx:
            mx = frq[i]
            ans = i
print(ans)

# input file:
# Python is easy to learn.
# Python is powerful and python is popular.
```
