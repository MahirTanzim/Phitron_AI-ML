## Read as a List of String.

```python
file = open('./sample_data/sample.txt', 'r')
content = file.readlines()     #readline-only read first line, readline read all lines in a list
print(content)
print(type(content))
file.close()
file.closed
```

## Open file using with keyward  (Best Practice)

```python
with open ("./sample_data/sample.txt", 'r') as file:
    content = file.readlines()
    print(content)
print(file.closed)
```

```python
with open ("./sample_data/sample.txt", 'r') as file:
    content = file.readlines()
    content = list(map(str.strip, content))    #.strip() is used to remove last space or othe element from string
    print(content)
    filter_cont = list(filter(lambda x : len(x) >13, content))
    print(filter_cont)
```

```python
with open ("./sample_data/sample.txt", 'r') as file:
    for line in file:
        print(line, end="")
print(file.closed)
```

```python
with open ("./sample_data/sample.txt", 'r') as file:
    for line in file:
        l = line.strip()
        print(l)
print(file.closed)
```

# File Pointer

## tell()

```python
with open('./sample_data/sample.txt', 'r') as file:
    print(file.tell()) #current cursor
    print(file.read())
    print(file.tell())
    print(file.read()) # no output. karon last index theke pora shuru korsi
```

We can modify read function  

```python
with open('./sample_data/sample.txt', 'r') as file:
    print(file.tell())
    print(file.read(19)) #read 19 element. it is not index. it is number of character to read
    print(file.tell())
    print(file.read(30))
    print(file.tell())
```

## Seek()



```python
with open('./sample_data/sample.txt', 'r') as file:
    print(file.tell())
    file.seek(10) # seek move the cursor to the given index (here in 10)
    print(file.tell())
    file.seek(5) # and here to index
    print(file.tell())
    print(file.read())
```

# Practice 1 :

Get the text from sample.txt , and count the following


Number of lines

Number of words

Number of characters

and save it to the counter_of_string.txt **file**

```python
from functools import reduce
import string
with open('./sample_data/sample.txt') as file:
    strings = file.readlines()
    print(strings)
    total_lines = len(strings)
    print(f"Total Lines: {total_lines}")  #total line koyta

    # total words
    number_of_word = list(map(lambda x : len(x.split()), strings))
    total_number_of_word = reduce(lambda x,y : x+y, number_of_word)
    # for _ in number_of_word:
    #     total_number_of_word+=_
    print(total_number_of_word)
    # cleaning process
    strings_list = list(map(str.strip, strings))
    print(strings_list)
    # delete space
    strings_list = list(map(lambda x : x.replace(" ", '') , strings_list))
    print(strings_list)

    # number of character
    number_of_char = list(map(lambda x: len(x), strings_list))
    print(number_of_char)

    total_number_of_char = reduce(lambda x, y: x+y, number_of_char)
    print(total_number_of_char)

with open('./sample_data/counter_of_string.txt', 'w') as file:
    file.write(f'Total Lines: {total_lines}\nTotal number of words: {total_number_of_word}\nTotal number of character: {total_number_of_char}')
```
